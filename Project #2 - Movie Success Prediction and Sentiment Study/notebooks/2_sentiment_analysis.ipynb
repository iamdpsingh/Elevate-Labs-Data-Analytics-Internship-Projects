```python
import pandas as pd
import matplotlib.pyplot as plt
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk
nltk.download('vader_lexicon')
```

```python
# Load reviews dataset (you'll need IMDB reviews dataset separately)
df = pd.read_csv('../data/IMDB Dataset.csv')  # Format: review, sentiment
```

```python
# Initialize VADER
sia = SentimentIntensityAnalyzer()
```

```python
# Add sentiment scores
df['compound'] = df['review'].apply(lambda x: sia.polarity_scores(str(x))['compound'])
```

```python
# Label sentiment
df['vader_sentiment'] = df['compound'].apply(lambda x: 'positive' if x >= 0.05 else 'negative' if x <= -0.05 else 'neutral')
```

```python
# Visualize
df['vader_sentiment'].value_counts().plot(kind='bar', title='VADER Sentiment Distribution')
plt.tight_layout()
plt.savefig('../visuals/sentiment_charts.png')
```
